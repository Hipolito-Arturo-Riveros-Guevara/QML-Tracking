```python
import pandas as pd
```

```python
res=pd.read_csv('resultado.csv')
tc1=pd.read_csv('TC_1.csv')
```

```python
l_0=res[res['label']==0].sample(n=2000, random_state=42)
l_1=tc1[tc1['label']==1].sample(n=2000, random_state=42)
df_unido = pd.concat([l_0, l_1], ignore_index=True)
df_mezclado = df_unido.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
df_mezclado.to_csv('sample1.csv', index=False)
```
